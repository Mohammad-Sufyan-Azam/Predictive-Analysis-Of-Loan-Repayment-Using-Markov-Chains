```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []
def find_max(ini_vec):
    
    arr_mat = np.array(
        [[0.47, 0.1, 0.16, 0.01, 0.26], [0.2, 0.32, 0.2, 0.22, 0.06], [0.3, 0.2, 0.38, 0.05, 0.07], [0, 0, 0, 1, 0],
         [0, 0, 0, 0, 1]])
    x = ini_vec
    i = 0
    while (True):
        ini_vec = x
        x = np.dot(ini_vec, arr_mat)
        i += 1
        if np.array_equal(ini_vec, x):
            break

    print("no of iterations for convergence ",i)
    return ini_vec


max_paid = 0.0
a1 = 0
b1 = 0
c1 = 0

for i in range(101):
    t = 100 - i
    b = i
    mx=0
    for j in range(t + 1):
        a = j
        c = t - j
        a = a / 100
        b = b / 100
        c = c / 100

        
        l = find_max(np.array([a, b, c, 0, 0]))
        mx=max(mx,l[4])
        if (l[4] > max_paid):
            a1 = a
            b1 = b
            c1 = c
            max_paid = l[4]
    x.append(i/100)
    y.append(mx)
    
print(max_paid)
print(a1)
print(b1)
print(c1)

print(x)
print(y)


# plotting the points
plt.plot(x,y, color="blue")

# naming the x axis
plt.xlabel('Initial probability of high risk')
# naming the y axis
plt.ylabel('probabilities of paid loan after n iterations ')

# giving a title to my graph
plt.title('Initial probability of being high risk loan')

# function to show the plot
plt.show()

```

```python
# when initial state is good loan

d1=find_max(np.array([1,0,0,0,0]))
print(d1)
a1=np.array(["Bad Loans","Paid up loans"])
b1=np.array([d1[3],d1[4]])

plt.bar(a1,b1,color='maroon',width=0.4)

plt.xlabel(" Transient states ")
plt.ylabel(" corrresponding probabilities of being at bad loans or paid up loans after convergence  ")
plt.title("Graph to show probabilities of bad and paidup loans when initial state is good loan ")
plt.show()
```
